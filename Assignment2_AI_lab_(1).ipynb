2. Consider the system of equations
3x + 2y = 5
10x − y = 3 (3)
Find the unique solution to these equations using matrix calculus in
Numpy array.

```python
import numpy as np

A = np.array([
    [3, 2],
    [10, -1]
])

b = np.array([5, 3])

solution = np.linalg.solve(A, b)

print("The solution is:")
print(f"x = {solution[0]}")
print(f"y = {solution[1]}")
```

3. Generate a 5 by 5 random matrix and find the maximum, minimum,
average (mean) & variance values along all axes.

```python
import numpy as np


random_matrix = np.random.rand(5, 5)


max_values = np.max(random_matrix, axis=0)
min_values = np.min(random_matrix, axis=0)
mean_values = np.mean(random_matrix, axis=0)
variance_values = np.var(random_matrix, axis=0)


print("Random Matrix (5x5):\n", random_matrix)


print("\nMaximum values along columns:\n", max_values)
print("Minimum values along columns:\n", min_values)
print("Mean values along columns:\n", mean_values)
print("Variance values along columns:\n", variance_values)
```

4. Generate a 5 by 5 random integer matrix whose values are between 2 &
100. Find the another matrix which produces boolean output as true
where Values of random matrix are between 20 & 70.

```python
import numpy as np


random_matrix = np.random.randint(2, 101, size=(5, 5))

boolean_matrix = (random_matrix >= 20) & (random_matrix <= 70)

print("Random Integer Matrix (5x5):\n", random_matrix)

print("\nBoolean Matrix (True where values are between 20 and 70):\n", boolean_matrix)
```

6. Create a Normalized random variable of size 20 using Numpy and find
out each quartile & their corresponding counts.

```python
import numpy as np

random_variable = np.random.randn(20)

quartiles = np.percentile(random_variable, [25, 50, 75])

counts = [
    np.sum(random_variable <= quartiles[0]),
    np.sum((random_variable > quartiles[0]) & (random_variable <= quartiles[1])),
    np.sum((random_variable > quartiles[1]) & (random_variable <= quartiles[2])),
    np.sum(random_variable > quartiles[2])
]

# Print the results
print("Random Variable:\n", random_variable)
print("\nQuartiles (Q1, Q2, Q3):\n", quartiles)
print("\nCounts in each quartile (Q1, Q2, Q3, Q4):\n", counts)
```

7. Find out the eigen value and eigen vectors of covariance of the data ma-
trix, X=[[4,11],[8,4],[13,5],[7,14]] where each row is one observation while

colums are attribures of that observation. (Hint: Question is related to
PCA & use np.cov and np.linalg.eig functions of numpy package)

```python
import numpy as np

X = np.array([[4, 11], [8, 4], [13, 5], [7, 14]])

cov_matrix = np.cov(X.T)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Print the results
print("Covariance Matrix:\n", cov_matrix)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
```

8. Create a pandas data structure (series) X using a dictionary D = "Rohan":
176,"John": 175,"Harvinder": 180,"Hasan": 174 . Then pass other list
["Tinku","John","Ramesh","Hasan"] as index into series X . Comment on
the resultant output & locate the index of NAN values.

```python
import pandas as pd

D = {"Rohan": 176, "John": 175, "Harvinder": 180, "Hasan": 174}

X = pd.Series(D)

X = X.reindex(["Tinku", "John", "Ramesh", "Hasan"])

print("Resultant Series X:\n", X)

nan_indices = X.index[X.isna()]

print("\nIndices of NaN values:", nan_indices.tolist())
```

10. Using the following data plot a bar plot and a horizontal bar plot.
company = ["Apple", "Microsoft", "Google", "AMD"]
profit = [3000, 8000, 1000, 10000]

```python
import matplotlib.pyplot as plt

company = ["Apple", "Microsoft", "Google", "AMD"]
profit = [3000, 8000, 1000, 10000]

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

axs[0].bar(company, profit, color='blue')
axs[0].set_title('Vertical Bar Plot')
axs[0].set_xlabel('Company')
axs[0].set_ylabel('Profit')

axs[1].barh(company, profit, color='green')
axs[1].set_title('Horizontal Bar Plot')
axs[1].set_xlabel('Profit')
axs[1].set_ylabel('Company')

plt.tight_layout()

plt.show()
```
